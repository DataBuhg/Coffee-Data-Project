```python
from sklearn.model_selection import train_test_split
sampled_df, _ = train_test_split(Df2, test_size=0.9, random_state=33) 
sampled_df = sampled_df.sort_values(by='Date')
```

```python
df2plot = sns.lineplot(x=sampled_df['Date'], y=sampled_df['Total Revenue Per Day'], linestyle='-', linewidth=2)
df2plot.set_title("Revenue Per Day")
df2plot.set_xlabel('Date')
df2plot.set_ylabel('Revenue Per Day')
plt.xticks(rotation=45)
plt.savefig('Revenue Per Day')
plt.show()
```

```python
labels = df3['Payment Type']
sizes = df3['Total Revenue']
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%', shadow=True, startangle=175)
plt.title('Percentage Of Payment Types')
plt.axis('equal')
plt.savefig("Percentage Of Payment Types")
plt.show()
```
